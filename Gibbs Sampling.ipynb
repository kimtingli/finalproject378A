```python
import random,math
 
def gibbs(N=50000,thin=1000):
    x=0
    y=0
    print("Iter",x,y)
    for i in range(N):
        for j in range(thin):
            x=random.gammavariate(3,1.0/(y*y+4))
            y=random.gauss(1.0/(x+1),1.0/math.sqrt(2*x+2))
        print(i,x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
def gibbs(n):
    #setting empty arrays for storing values later
    x = np.empty(n)
    y = np.empty(n)
    a = np.empty([n,2])
    #setting the initial values for x and y, a is simply a matrix form of x,y
    x[0] = 5
    y[0] = 5
    a[0] = (5,5)
    for i in range(1,n):
        #sample from f(x|y) using inversion method
        u1 = np.random.uniform(0,1)
        x[i] = np.log(1-u1)/(-y[i-1])
        #sample from f(y|x) using inversion method
        u2 = np.random.uniform(0,1)
        y[i] = np.log(1-u2)/(-x[i])
        a[i] = (x[i],y[i])
    print(a)
    #histogram of X
    plt.hist(x, 500, normed=1, facecolor='blue', alpha=0.5)
    plt.show()
```

```python
gibbs(500)
```
